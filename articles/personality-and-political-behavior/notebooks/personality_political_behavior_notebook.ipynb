# Personality and Political Behavior

This notebook scaffold demonstrates how to load the synthetic dataset, summarize political-personality patterns, and examine ideology, participation, identity-threat, and institutional trust.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_political_behavior.csv')
df.head()
```

```python
metrics = ['ideology_score', 'political_participation', 'affective_polarization', 'trust_in_institutions']
df.groupby('country_context')[metrics].mean().round(3)
```
