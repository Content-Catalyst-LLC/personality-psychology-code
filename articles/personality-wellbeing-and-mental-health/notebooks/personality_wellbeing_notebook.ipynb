# Personality, Wellbeing, and Mental Health

This notebook scaffold demonstrates how to load the synthetic dataset, summarize wellbeing patterns, and examine distress, flourishing, support, coping, and two-continua mental-health profiles.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_wellbeing_mental_health.csv')
df.head()
```

```python
metrics = ['wellbeing_score', 'distress_score', 'flourishing_score', 'perceived_support', 'stress_burden']
df.groupby('life_context')[metrics].mean().round(3)
```
