# Personality, Work, and Leadership

This notebook scaffold demonstrates how to load the synthetic dataset, summarize work-personality patterns, and examine performance, leadership emergence, leadership effectiveness, teamwork, and derailment risk.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_work_leadership.csv')
df.head()
```

```python
metrics = ['job_performance', 'leadership_emergence', 'leadership_effectiveness', 'teamwork_quality', 'burnout_risk']
df.groupby('role_family')[metrics].mean().round(3)
```
