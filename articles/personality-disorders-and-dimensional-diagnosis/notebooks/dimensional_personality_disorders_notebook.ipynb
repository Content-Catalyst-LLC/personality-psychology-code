# Personality Disorders and Dimensional Diagnosis

This notebook scaffold demonstrates how to load the synthetic dataset, summarize dimensional personality-pathology patterns, and examine severity, maladaptive trait domains, functioning impairment, risk, and treatment engagement.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_disorders_dimensional_diagnosis.csv')
df.head()
```

```python
metrics = ['functioning_impairment', 'maladaptive_trait_burden', 'pd_severity', 'risk_level', 'treatment_engagement']
df.groupby('clinical_context')[metrics].mean().round(3)
```
