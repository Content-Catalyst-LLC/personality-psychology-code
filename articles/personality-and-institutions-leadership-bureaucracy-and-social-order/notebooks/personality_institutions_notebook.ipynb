# Personality and Institutions

This notebook scaffold demonstrates how to load the synthetic dataset, summarize institutional-personality patterns, and examine role fit, discretion, accountability, and institutional trust.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_institutions_bureaucracy.csv')
df.head()
```

```python
metrics = ['bureaucratic_fit', 'discretion_level', 'accountability_strength', 'institutional_performance', 'institutional_trust']
df.groupby('institutional_unit')[metrics].mean().round(3)
```
