# Personality, Relationships, and Social Functioning

This notebook scaffold demonstrates how to load the synthetic dataset, summarize relational-personality patterns, and examine relationship satisfaction, social functioning, loneliness, conflict, and perceived support.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_relationships_social_functioning.csv')
df.head()
```

```python
metrics = ['relationship_satisfaction', 'social_functioning', 'loneliness', 'conflict_frequency', 'perceived_support']
df.groupby('social_context')[metrics].mean().round(3)
```
