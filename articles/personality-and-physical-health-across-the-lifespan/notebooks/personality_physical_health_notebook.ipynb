# Personality and Physical Health Across the Lifespan

This notebook scaffold demonstrates how to load the synthetic dataset, summarize lifespan personality-health patterns, and examine physical health, functional ability, health behavior, stress burden, and chronic-condition burden.

```python
from pathlib import Path
import pandas as pd

root = Path('..')
df = pd.read_csv(root / 'data' / 'synthetic_personality_physical_health_lifespan.csv')
df.head()
```

```python
metrics = ['physical_health_score', 'functional_ability', 'chronic_condition_burden', 'stress_burden']
df.groupby('age_band')[metrics].mean().round(3)
```
